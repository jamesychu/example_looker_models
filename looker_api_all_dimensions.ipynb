```python
import numpy as np
```

```python
all_models = sdk.all_lookml_models()
```

```python
nl = []
for i in all_models:
    if i['project_name']=='rsi_snowflake':
        nl.append(i)
```

```python
model_name = []
explore_name = []

for i in range(len(nl)):
    model_name.extend([nl[i]['name']].copy() * len(nl[i]['explores']))
    for j in range(len(nl[i]['explores'])):
        explore_name.append(nl[i]['explores'][j]['name'])
        
```

```python
sdk.lookml_model_explore('RSI_SF','kpi_player')
```

```python
dim_names = []
view_name_list = []
model_name_list = []
explore_name_list = []

for i, j in zip(model_name,explore_name):
    
    t_string = sdk.lookml_model_explore(lookml_model_name = i  , explore_name = j)['fields']['dimensions']
    model_name_list.extend([i].copy() * len(t_string))
    explore_name_list.extend([j].copy() * len(t_string))
    for x in range(len(t_string)):
        dim_names.append(t_string[x]['name'])
    ##    model_name_list.extend([i].copy() * len(t_string))
      ##  explore_name_list.extend([j].copy() * len(t_string))
        
```

```python
import pandas as pd
```

```python
looker_dimensions = pd.DataFrame({'Model_Name': model_name_list,
                                  'Explore_Name': explore_name_list,
                                   'Dimension': dim_names})
```

```python
looker_dimensions.to_csv('looker_rsi_sf_dimensions_withoutr.csv')
```
